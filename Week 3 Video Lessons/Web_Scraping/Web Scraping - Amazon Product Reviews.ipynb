## Web Scraping Using BeautifulSoup in Python – Scraping Amazon Product Reviews using Python

```python
# importing the pakages 
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = 'https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
```

```python
page
```

```python
# .content function returns the raw bytes of the response content.
page.content
```

```python
# Creating the beautiful soup object
soup = bs(page.content,'html.parser')
```

```python
# The prettify() method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string.
print(soup.prettify())
```

```python
# find() is a Python library function which is used to find the index of first occurence of a substring from the given string. 
names = soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
# Here we are creating an empty list and storing the names inside
# The append() method in python adds a single item to the existing list. 
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
# Python list pop() is an inbuilt function in Python that removes and returns the last value from the List or the given index value.
cust_name.pop(0)
```

```python
# Here it returns the list without the first name
cust_name
```

```python
# find_all() returns all the tags and strings that match your filters and this can take a while if the document is large.
title = soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
# creating a loop and storing the title in to an empty list
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
# The lstrip() method returns a copy of the string with leading characters removed
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
#The rstrip() method removes any trailing characters (characters at the end of a string), space is the default trailing character to remove.
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
# finding by class 
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
# using a loop creating a list of rating and storing it in the list.
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
# the len() returns the length of the given string, array, list, tuple, dictionary
len(rate)
```

```python
# List pop in Python is a pre-defined, in-built function that removes an item at the specified index from the list.
rate.pop(0)
```

```python
rate.pop(0)
```

```python
rate
```

```python
# Here we will apply the find_all method using the span tag 
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
# And also store it in an empty list
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
# Here we will use the lstrip ie removing all the string characters starting from the left
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
# the rstrip ie removing all the string characters starting from the right
review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
# lets check the lenght of review_content
len(review_content)
```

```python
#the .dataframe method is use to inserts a column into a DataFrame
df = pd.DataFrame()
```

```python
df
```

```python
# So here lets create a dataframe from the customer name column
df['Customer Name']=cust_name
```

## Scraping HTML tables with the help of pandas

## Setup

```python
# Importing the necessary libraries
import requests
from bs4 import BeautifulSoup
```

```python
# We define the url
base_site=("https://en.wikipedia.org/wiki/List_of_national_capitals_by_population")
```

```python
# Here we make a get request
r = requests.get(base_site)
```

```python
# finally we extract the html of the page from the response
html=r.content
```

### Extracting tables with Beautiful Soup

```python
# First we parse the html in the constructor along with the lxml parser
soup= BeautifulSoup(html, "lxml")
```

```python
# Now we can search for all the tables in the document
soup.find_all("table")[1]
```

```python
# Lets store it in a variable for further manipulation
table = soup.find_all("table")[1]
```

```python
# Different tags in a table
# th: Table Heading (name of the column)
# tr: Table Row
# td: Table Data Cell
```

```python
# its a good idea to inspect the first content of the first row of this table
table.find_all("tr")[0].contents
```

```python
# note: the above returns a new row as a different element in the list
# thus to achieve our goals of scraping the capital cities we need to take the 6th element of every row, and we will extract this with a simple list comprehension 
# we will be looking for the string of each element with index 5 for each row from the table
capitals=[row.contents[3].text for row in table.find_all("tr")]
capitals
```

 So far we have mannually inspected the structure of the table and extracted only one column from it, at this point you probably realize how tedious this process can can become. Not only do we have to repeat each process for every column but the structure has lost some of its initial tabular form, and we need to reconstruct it.That is where you probably start to wonder if there is a better way to do this and its safe to say you already know the answer to this question since the lecture is about web scrapping html tables.
