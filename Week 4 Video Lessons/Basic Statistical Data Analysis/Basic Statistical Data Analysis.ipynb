### Shapiro-Wilks Test

##### Test for normality

The Shapiro-Wilks test for normality is one of three general normality tests designed to detect all departures from normality.  It is comparable in power to the other two tests.

The test rejects the hypothesis of normality when the p-value is less than or equal to 0.05.  Failing the normality test allows you to state with 95% confidence the data does not fit the normal distribution.  Passing the normality test only allows you to state no significant departure from normality was found.

#### H0: The Data are normally distributed

##### Accept H0 is p>0.05

```python
shapiro_results = scipy.stats.shapiro(sl)
```

```python
print(shapiro_results) #test statistic and p-value
```

### Standard Normal Distribution

```python
# Importing required package
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python
mu = 0 # mean value
variance = 1 # variance in measured data
sigma = math.sqrt(variance) # # standard deviation of measured data
# Generate x-axis with 3 standard deviations to the left of the mean and 3 to the right of it
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100) 
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

##### Calculate Confidence Level

point estimates are estimates of population parameters based on sample data. If we wanted to know the everage weights of students in a school, we will a collect a sample to estimate the average weight of all students in a school

The sample mean is usually not exactly the same as population mean

A confidence interval is a range of values above and below a point estimate that captures the true population paramater at some predetermine confidence level

If you want to have a 95% chance of capturing the true population parameter with a point estimates and a corresponding confidence interval,you'd set your confidence level to 95%

##### CL = mean  margin errror

##### Where MOE is Z*sd/sqrt(n)

```python
# importing required packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math
import random
import scipy.stats as stats
```

```python
# We'll set a seed here so our runs are consistent
np.random.seed(10)
population_wt = stats.poisson.rvs(loc=18,mu=35, size=150000) #randomly creates population weight data for 150k people

sample_size= 1000 
sample = np.random.choice(a=population_wt, size=sample_size) 
sample_mean = sample.mean() #samples to 1000 people

z_critical = stats.norm.ppf(q = 0.975) # Get the z-critical value at 95% level*
# z-value at 95% CI is 1.96

print('Z-critical value:')            # check the z-critical value
print(z_critical)

pop_stdev = population_wt.std()           #Get the population standard deviation

margin_of_error= z_critical * (pop_stdev/math.sqrt(sample_size))

confidence_interval = (sample_mean - margin_of_error,
                      sample_mean + margin_of_error)

print('confidence interval:')
print(confidence_interval)
```
