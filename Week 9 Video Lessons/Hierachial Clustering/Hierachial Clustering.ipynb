# Python Training Hierachical Clustering Explained

Hierarchical clustering, also known as hierarchical cluster analysis, is an algorithm that groups similar objects into groups called clusters. The endpoint is a set of clusters, where each cluster is distinct from each other cluster, and the objects within each cluster are broadly similar to each other.

```python
# importing the data set
dataset = pd.read_csv('C:\\Users\\Zachary_B\\Desktop\\dataset\\Wholesale+customers+data.csv')
```

```python
# feature scaling
data_scaled = normalize(dataset)
data_scaled = pd.DataFrame(data_scaled,columns = dataset.columns)
data_scaled.head()
X = data_scaled
X.head()
```

```python
# using the dendogram to find the optimal number of clusters
dendogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

## Python Training Hierarchical Clustering

```python
# loading iris dataset
iris = datasets.load_iris()
```

```python
# seperating our independent variable and extracting the 1st 2 columns from the data set into it
X = iris.data[:20,:2]
y_iris = iris.target
```

```python
# creating an agglomerate cluster and ploting a scatter plot
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
plt.scatter(X[y_hc == 0,0], X[y_hc == 0,1], s=10, c = 'red', label = 'Cluster1')
plt.show()
```

```python
# 
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
plt.scatter(X[y_hc == 0,0], X[y_hc == 0,1], s=10, c = 'red', label = 'Cluster1')
plt.scatter(X[y_hc == 1,0], X[y_hc == 1,1], s=10, c = 'blue', label = 'Cluster2')
plt.show()
```

```python
# creating a dendrogram base on the clusters
dendogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.show()
```
