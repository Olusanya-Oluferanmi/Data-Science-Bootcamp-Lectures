```python
# Loading the library with the iris dataset
from sklearn.datasets import load_iris
# Loading scikit'srandom forest classifier library
from sklearn.ensemble import RandomForestClassifier
# Loading pandas
import pandas as pd
# Loading numpy
import numpy as np
# Setting random seed
np.random.seed(0)
```

```python
# Creating an object called iris with the iris data
iris = load_iris()
# Creating a dataframe with the four feature variables
df=pd.DataFrame(iris.data, columns=iris.feature_names)
# Viewing the top 5 rows
df.head()
```

```python
print(iris)
```

```python
# Adding a new column for the species name
df['species']= pd.Categorical.from_codes(iris.target,
                                        iris.target_names)
# Viewing the top 5 rows
df.head()
```

```python
# Creating Test and train Data
df['is_train'] = np.random.uniform(0,1,len(df)) <= .75
# View the top 5 rows
df.head()
```

```python
# Creating dataframes with test rows and training rows
train, test = df[df['is_train']==True], df[df['is_train']==False]
# Show the number of observations for the test and training data frames
print('Number of observation in the training data:', len(train))
print('Number of observation in the test data:', len(test))
```

```python
# Create a list of the feature column's names
features=df.columns[:4]
# View features
print(features)
```

```python
# Converting each species name into digits
y= pd.factorize(train['species'])[0]
# Viewing target
y
```

```python
# Creating a random forest classifier.
clf= RandomForestClassifier(n_jobs=2, random_state=0)

# Training the classifier
clf.fit(train[features],y)
```

```python
# Applying the trained classifier to the test
clf.predict(test[features])
```
