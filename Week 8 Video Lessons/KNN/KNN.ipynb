```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
# Importing the dataset
dataset = pd.read_csv('C:\\Users\\Feranmi\\Downloads\\diabetes.csv')
```

```python
print(len(dataset))#number of rows
print(dataset.head())#overview of the dataset
```

```python
# replace zeros
zero_not_accepted=['Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'Insulin']

for column in zero_not_accepted:
    dataset[column] = dataset[column].replace(0, np.NaN)
    mean = int(dataset[column].mean(skipna = True))
    dataset[column] = dataset[column].replace(np.NaN, mean)
```

```python
print(dataset['Glucose'])
```

```python
#split dataset
X = dataset.iloc[:, 0:8]
y = dataset.iloc[:, 8]
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0, test_size=0.2)
```

```python
#Feature scaling
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import math
math.sqrt(len(y_test)) #sqrt- square root
```

```python
# Define the model: Init k-NN
classifier = KNeighborsClassifier(n_neighbors=11, p=2, metric='euclidean')
```

```python
# fit model
classifier.fit(X_train, y_train)
```

```python
# Predict test set result
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Evaluate model
cm= confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print(f1_score(y_test, y_pred)) #is a measure of a test's accuracy.It is a blend of the precision and recall
```

```python
print(accuracy_score(y_test, y_pred))#is used to calculate the accuracy of either the faction or count of correct prediction
```
