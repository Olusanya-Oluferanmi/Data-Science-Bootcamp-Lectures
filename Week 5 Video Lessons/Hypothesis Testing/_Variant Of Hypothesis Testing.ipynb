## Correlation of categorical variable - Chi square test

```python
#Test wheather two categorical variables are related or independent
# Assumptions - independent observation, size in each body of contingency table > 25
# Example of the chi-Squared Test
```

```python
loandata=pd.read_csv('')
```

```python
loandata.head()
```

```python
contingency_data= pd.crosstab(loandata['purpose'], loandata['loan_status'], margins=False)
```

```python
contingency_data
```

```python
from scipy.stats import chi2_contingency
```

```python
stat, p, dof, expected = chi2_contingency(contingency_data)
print('stat=%.3f, p=%.5f'%(stat,p))

if p> 0.05:
    print('Independent categories')
else:
    print('dependent categories')
```
