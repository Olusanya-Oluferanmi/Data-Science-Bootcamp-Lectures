## Maximum Likelihood Estimation With Simple Examples And Python

```python
# importing the libraries
import pandas as pd
import numpy as np
import math
```

```python
# loading the sample data
x=np.array([[1, 2, 3, 4, 5]])
```

```python
# checking the shape
n=np.shape(x)[1]
```

```python
# print the sample data
print('The Sample data: \np',x)
print('The number of data present in x is: ',np)
```

```python
# the combination of mu and sigma
mu= np.array([[1, 2, 3, 2, 3]])
sigma=np.array([[1, 2, 2, 3, 3]])
print('\npThe mean is: ',mu)
print('The st.dev is: ',sigma)
```

```python
# calculating the log for each and every set
print('Mu Sigma logL')
for i in range(5):
    logL = - np.sum(np.square(x - mu[0,i])/(2*np.square(sigma))) -(0.5*n*np.log10(2*math.pi)) - n*np.log10(sigma[0,i])
    print(mu[0,i], sigma[0,i], np.round(logL,3))
```
