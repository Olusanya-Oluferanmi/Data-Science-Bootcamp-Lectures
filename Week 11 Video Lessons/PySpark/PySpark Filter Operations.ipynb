```python
# Import session
from pyspark.sql import SparkSession
```

```python
Spark= SparkSession.builder.appName('dataframe0000000000000').getOrCreate()
```

```python
# Read csv file
df_pyspark=spark.read.csv('test1pyspark.csv', header=True, inferSchema=True)
df_pyspark.show()
```

## Filter Operations

```python
#salary of the people less than or equal to 20000
df_pyspark.filter('Salary<=20000').show()
```

```python
# Name and age of people less than or equal to 20000
df_pyspark.filter('Salary<=20000').select(['Name','age']).show()
```

```python
#Method 2: salary of the people less than or equal to 20000
df_pyspark.filter(df_pyspark['Salary']<=20000).show()
```

```python
# More conditions
df_pyspark.filter((df_pyspark['Salary']<=20000)&
                  (df_pyspark['Salary']>=15000)).show()
```

```python
# Not operation(~)
df_pyspark.filter(~(df_pyspark['Salary']<=20000)).show()
```
