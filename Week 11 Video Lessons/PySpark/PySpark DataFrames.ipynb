```python
# import sparksession
from pyspark.sql import SparkSession
```

```python
# start session
spark = SparkSession.builder.appName('Dataframe').getOrCreate()
```

```python
spark 
```

```python
# Read the dataset: method 1
df_pyspark=spark.read.option('header', 'true').csv('test1pyspark.csv', inferSchema=True).show()
# if you don't indicate inferSchema as true it will recognize all 
#the values as string 
# If you don't remove the .show() you'll get an error while 
#checking for the schema because it has being put into vaviable pyspark thus affecting the outcome
```

```python
# Check the schema(datatype)
df_pyspark.printSchema()
```

```python
# Read the dataset: method 2
df_pyspark=spark.read.csv('test1pyspark.csv', header=True, inferSchema=True)
df_pyspark.show()
```

```python
# Find data type
type(df_pyspark)
# dataframe is a data structure where you can perform various operations
# types: pyspark, pandas
```

```python
#Selecting Columns 
df_pyspark.head()
# it's output is usually in a list format
```

```python
# returns it in a dataframe format
df_pyspark.show()
```

```python
# How to pick up columns
df_pyspark.select('Name').show()
```

```python
# returns datatype
type(df_pyspark.select('Name'))
```

```python
# How to pick up multiple columns
df_pyspark.select(['Name', 'Experience']).show()
```

```python
# returns datatype of all the columns
df_pyspark.dtypes
```

```python
## Check Describe Option Similar To Pandas
df_pyspark.describe().show()
# it describes string columns too
```

```python
#Adding Columns in dataframe
df_pyspark.withColumn('Experience After 2 years',df_pyspark['Experience']+2 )
```

```python
df_pyspark.withColumn('Experience After 2 years',df_pyspark['Experience']+2 ).show()
```

```python
df_pyspark=df_pyspark.withColumn('Experience After 2 years',df_pyspark['Experience']+2 )
df_pyspark.show()
```

```python
# Drop the columns
df_pyspark=df_pyspark.drop('Experience After 2 years')
```

```python
df_pyspark.show()
```

```python
## Rename the columns
df_pyspark.withColumnRenamed('Name','New Name').show()
```
