Pyspark Handling Missing Values

- Dropping Columns
- Dropping Rows
- Various Parameter in Dropping Functionalities
- Handling Missing Values by Mean, Median And Mode

```python
#Import spark session
from pyspark.sql import SparkSession
Spark= SparkSession.builder.appName('Practise').getOrCreate()
```

```python
#Read Dataset
df_pyspark=spark.read.csv('test2.csv',header=True, inferSchema=True)
```

```python
# Drop the column
```

```python
df_pyspark.drop('Name').show()
```

```python
# deletes all rows with null values by default
df_pyspark.na.drop().show()
```

```python
# Dropping "all" the null rows
# any==how
df_pyspark.na.drop(how='all').show()
```

```python
df_pyspark.na.drop(how='any').show()
```

```python
# threshhold
# This takes int value, Drop rows that have less than thresh hold n
# on-null values. Default is ‘None’.

df_pyspark.na.drop(how='any', thresh=2).show()
```

```python
#Subset: This is when you want to drop null values in the columns
df_pyspark.na.drop(how='any', subset=['Experience']).show()
```

```python
## Filling the missing values
df_pyspark.na.fill('Missing Values', 'Experience').show()
```

```python
df_pyspark.na.fill('Missing Values', ['Experience','age']).show()
```

```python
# replace null values in specified columns with the mean
from pyspark.ml.feature import Imputer

imputer=Imputer(
    inputCols=['age','Experience', 'Salary'],
    outputCols=['{}_imputed'.format(c) for c in ['age', 'Experience', 'Salary']]
    ).setStrategy('mean')
```

```python
# Add imputation columns to dataframe
imputer.fit(df_pyspark).transform(df_pyspark).show()
```
