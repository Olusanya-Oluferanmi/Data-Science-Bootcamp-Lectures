### Symbols for writing regular expresssion
*: The preceding character is repeated zero or more times

+: The preceding character is repeated at least once

{}: The preceding character is repeated as many times as mentioned in the braces

.: Represents a single occurrence of any character except newline

?: The preceding character is optional

^: Specifies that the match must start at the begining of the string

$: Specifies that the match must occur at the end of the string

[]: Matches one out the all characters within the brackets

[^..] : Matches any one character excepts those note in thhe brackets

\d: matches a digit

\w: matches an alphanumeric character

\s: matches a whitespace character


```python
str='Abcd 4 computer 765 Python 687'
pattern = 'computer'# extracting a specific word
match=re.findall(pattern,str)
print(match)
```

```python
str='Abcd 4 computer 765 Python 687'
# Returns aplhabeth a-zA-Z
pattern = r'[a-zA-Z]+' 
match=re.findall(pattern,str)
print(match)
```

```python
str='Abcd 4 computer 765 Python 687'
# Returns numbers
pattern = r'[0-9]+'
match=re.findall(pattern,str)
print(match)
```

```python
str='Abcd 4 computer 765 Python 687'
# Returns aplhabeth and numbers
pattern = r'[a-zA-Z0-9]+'
match=re.findall(pattern,str)
print(match)
```

```python
str='Abcd 4 computer 765 Python 687'

pattern = r'.[^ ]+'
match=re.findall(pattern,str)
print(match)
```

```python
str='''
apple
banana
orange
peach
avocado
cherries
'''

pattern= r'.*s'
match=re.findall(pattern,str)

for m in match:
    print(m)
```

```python
str='''
apple
banana
orange
peach
avocado
cherries
'''
# returns words with vowel Word boundaries
pattern= r'\b[aeiou].+\b'
match=re.findall(pattern,str)

for m in match:
    print(m)
```

```python
str='''
apple
banana
orange
peach
avocado
cherries
'''
# returns words with vowel
pattern= r'[aeiou].+'
match=re.findall(pattern,str)

for m in match:
    print(m)
```

```python
str='''
olu@gmail.com
ara@yahoo.com 
zara@hotmail.com
ye98ji9.18
'''
# returns genuiue mail address
pattern= r'[a-z]+[0-9]*[a-z]*@[a-z]+\.com'
match=re.findall(pattern,str)

for m in match:
    print(m)
```

```python
str='''
olu@gmail.com
ara@yahoo.com 
zara@hotmail.com
ye98ji9.18
'''
# returns more information on mail address through the iteration
pattern= r'[a-z]+[0-9]*[a-z]*@[a-z]+\.com'
match=re.finditer(pattern,str)

for m in match:
    print(m)
```
