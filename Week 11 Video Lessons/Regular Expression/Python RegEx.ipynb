# Python RegEx

```python
# Imort regular expression library
import re
```

```python
Nameage='''
Janice is 22 and Theo is 33
Gabe is 44 and Joe is 21
'''
```

```python
# Condition to extract ages and names
ages= re.findall(r'\d{1,3}', Nameage)
names = re.findall(r'[A-Z][a-z]*', Nameage)
```

```python
# define variable
ageDict = {}
```

```python
# define variable
x=0
```

```python
# We'll extract all the names and ages into a dictionary
for eachname in names:
    ageDict[eachname] = ages[x]
    x+=1
```

```python
print(ageDict)
```

# Operations

```python
# find a word in a string
allinform = re.findall('inform', 'we need to inform him with the latest information')
for i in allinform:
    print(i)
# we have two outcomes beacause there's inform in 'information'
```

```python
if re.search('inform','we need to inform him the latest info'):
    print('There is inform')
```

```python
#generate an iterator: getting the starting and ending index of a particular string
str = 'we need to inform him the latest info'

for i in re.finditer('inform', str):
    
    loctup = i.span()
    
    print(loctup)
```

```python
# match one of any of several letters: match words with a particular pattern
Str= 'Sat,hat,mat,pat'
allStr = re.findall('[shmp]at',Str)

for i in allStr:
    print(i)
# 'Sat' is missing because it's case sensitive
```

```python
# match series of range of characters
Str = 'Sat,hat,mat,pat'
someStr = re.findall('[h-m]at', Str)

for i in someStr: 
    print(i)
```

```python
# replace a string
Food ='hat,hat,rat,pat'
 
regex =re.compile('[r]at')

Food =regex.sub('food', Food)

print(Food)
```

```python
randstr = 'hear is \\drogba'
print(randstr)
```

```python
randstr = 'hear is \\drogba'
print(re.search(r'\\drogba', randstr))
```

```python
# match a single character 
randStr='12345'
print('Matches:', len(re.findall('\d(5)',randStr)))
```

```python
# We replaced the new line with space: we have the three lines a single line 
randstr='''Keep the blue flag
flying high
chelsea'''

print(randstr)
regex = re.compile('\n')

randstr = regex.sub(' ',randstr)
print(randstr)
```

```python
# \b  backspace
# \f  formfeed
# \r:  carriage return
# \t  tab
# \v  vertical tab
# \w  [a-zA-Z0-9_] # small letter w will match all the things only in the bracket
# \W  [^a-zA-Z0-9]# capital W will match anything except the one in the bracket
# \n   new line
# \s  [\f\n\r\t\v]
# \s  [^\f\n\r\t\v] # It will include anything but this
```

```python
# \D will match anything except numbers but \d will match any number
num= '123 1234 12345 123456 1234567'

print('Matches:', len(re.findall('\d{5,7}', num)))
```

```python
No ='000-222-3333'# instead of \w you can use \d

if re.search('\w{3}-\w{3}-\w{4}', No):
    print('it is a phone no')
```

```python
\s -- same or equal to [\f\n\r\t\v]
```

```python
if re.search('\w{2,20}\s\w{2,20}', 'Olu Feranmi'):
    print('Fullname is valid')
```

```python
email = 'olu@gmail.com ara@yahoo.com zara@hotmail.com'

print('EmailMatches:', len(re.findall('[\w._%+-]{1,20}@[\w.-]{2,20}.[A-Za-z]{2,3}', email)))
```

```python
import urllib.request
from re import findall

url ='http://www.summet.com/dmsi/html/codesamples/addresses.html'

response = urllib.request.urlopen(url)

html= response.read()

htmlStr = html.decode()
pdata = findall('\(\d{3}\) \d{3}-\d{4}', htmlStr)

for item in pdata:
    print(item)
```
